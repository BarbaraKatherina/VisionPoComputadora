```python
import cv2
import numpy as np

def white_patch(image):
    max_rgb = np.max(image, axis=(0, 1), keepdims=True)
    corrected_img = (image / max_rgb) * 255.0
    return np.clip(corrected_img, 0, 255).astype(np.uint8)

# Leer la imagen en color
image_color = cv2.imread('wp_red2.jpg')

# Convertir la imagen a escala de grises
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

# Aplicar White Patch a la imagen en escala de grises
corrected_gray = white_patch(image_gray)

# Aplicar White Patch a los canales de color por separado
corrected_channels = [white_patch(image_color[:, :, i]) for i in range(3)]
corrected_color = np.stack(corrected_channels, axis=-1)

# Mostrar las imágenes originales y corregidas
cv2.imshow('Imagen en escala de grises original', image_gray)
cv2.imshow('Imagen en escala de grises corregida', corrected_gray)
cv2.imshow('Imagen en color original', image_color)
cv2.imshow('Imagen en color corregida', corrected_color)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Visualiza las dos imágenes
plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Imagen 1')

plt.subplot(1, 2, 2)
plt.imshow(corrected_gray, cmap='gray')
plt.title('Imagen 2')

plt.show()

# Visualiza las dos imágenes
plt.subplot(1, 2, 1)
plt.imshow(image_color)
plt.title('Imagen 1')

plt.subplot(1, 2, 2)
plt.imshow(corrected_color, cmap='jet')
plt.title('Imagen 2')

plt.show()

```
